```python
plt.hist(df.loc[df['Class'] == 1, 'V1'], bins = 20, density = True, color = ['red'], alpha = 0.5)
plt.hist(df.loc[df['Class'] == 0, 'V1'], bins = 20, density = True, color = ['grey'], alpha = 0.5)
plt.legend (['Class 1', 'Class 0'])
plt.xlabel('Class');
```
