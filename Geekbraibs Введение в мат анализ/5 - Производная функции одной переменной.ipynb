__2.__ Найти производную $y'_{x}$ функции:

### \begin{cases}
   y=\frac{t^2}{t-1}, \\
   x=\frac{t}{t^2-1} 
 \end{cases}

```python
t = Symbol('t')
y = t ** 2 / (t - 1)
y
```

```python
x = t / (t ** 2 - 1)
x
```

```python
diff(y, t) / diff(x, t)
```

__3.__ Найти производную с помощью логарифмирования:

### $$y=(x^2+2)^5\cdot(3x-x^3)^3$$

```python
x = Symbol('x')
y = (x ** 2 + 2) ** 5 * (3 * x - x ** 3) ** 3
y
```

```python
diff(y,x)
```

__4.__ Найти производную функции с помощью логарифмирования:

### $$y=x^x$$

```python
x = Symbol('x')
y = x ** x
y
```
