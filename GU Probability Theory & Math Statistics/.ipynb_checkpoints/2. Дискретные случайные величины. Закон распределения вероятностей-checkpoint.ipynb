```python
import numpy as np
from math import factorial
```

```python
# Формула сочетания
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

**Задание 1**  
Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

$$P_n(X=k) = C_n^k p^k q^{n-k}$$

```python
# Формула Бернулли
combinations(100, 85) * (0.8 ** 85) * (0.2 ** (100 - 85))
```

**Задание 2**  
Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек.  
а) Какова вероятность, что ни одна из них не перегорит в первый день?

$$P_m \approx \frac{\lambda^m}{m!}e^{-\lambda}$$

где $λ$ = $np$.

```python
# Формула Пуассона
n = 5000
m = 0
p = 0.0004
lambda_ = n * p

(lambda_**m / np.math.factorial(m)) * np.exp(-lambda_)
```

б) Какова вероятность, что перегорят ровно две?

```python
# Формула Пуассона
n = 5000
m = 2
p = 0.0004
lambda_ = n * p

(lambda_**m / np.math.factorial(m)) * np.exp(-lambda_)
```

```python
# Формула Бернулли
combinations(5000, 2) * (0.0004 ** 2) * ((1 - 0.0004) ** (5000-2))
```

**Задание 3**   
Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

```python
# Формула Бернулли
combinations(144, 70) * (0.5 ** 70) * (0.5 ** (144 - 70))
```

*Вариант 1*

а) Какова вероятность того, что **все мячи белые?**  

```python
# Первый ящик
combinations(7, 2) / combinations(10, 2)
```

```python
# Второй ящик
combinations(9, 2) / combinations(11, 2)
```

б) Какова вероятность того, что **ровно два мяча белые?**  

```python
p1 = combinations(7, 2) / combinations(10, 2) * combinations(2, 2) / combinations(11, 2) # 2 белых из первого
p1
```

```python
p2 = combinations(3, 2) / combinations(10, 2) * combinations(9, 2) / combinations(11, 2)  # 2 белых из второго
p2
```

```python
# 1 белый из первого и 1 белый из второго
p3 = (combinations(7, 1) * combinations(3, 1) / combinations(10, 2)) * (combinations(9, 1) * combinations(2, 1) / combinations(11, 2)) 
p3 
```

```python
p = p1 + p2 + p3
p
```
